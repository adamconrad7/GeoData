```python
import pandas as pd
import glob
import os
from os import listdir
from tqdm import tqdm
```

```python
headers='WBANNO UTC_DATE UTC_TIME LST_DATE LST_TIME CRX_VN LONGITUDE LATITUDE AIR_TEMPERATURE PRECIPITATION SOLAR_RADIATION SR_FLAG SURFACE_TEMPERATURE ST_TYPE ST_FLAG RELATIVE_HUMIDITY RH_FLAG SOIL_MOISTURE_5 SOIL_TEMPERATURE_5 WETNESS WET_FLAG WIND_1_5 WIND_FLAG'
headers = headers.split(' ')

agg= []
dir_path = 'CRNS0101-05-202309180550/2023'
start_date = pd.Timestamp('2023-09-11T15:10:00')
for fname in tqdm(listdir(dir_path)):
    df = pd.read_fwf(dir_path + '/' + fname, names= headers)
    df['UTC_DATE'] = pd.to_numeric(df['UTC_DATE'], downcast='integer', errors='coerce')
    df.dropna(inplace=True)
    df['UTC_DATE'] = df['UTC_DATE'].astype(int)
    df['UTC_TIME'] = df['UTC_TIME'].astype(float)
    df['UTC_TIME'] = df['UTC_TIME'].astype(int)
    df['UTC_TIME'] = df['UTC_TIME'].apply('{0:0>4}'.format)
    df['DATETIME'] = pd.to_datetime(df['UTC_DATE'].astype(str)  + df['UTC_TIME'].astype(str), format='%Y%m%d%H%M')
    df = df[df['DATETIME'] >= start_date]
    cols_to_order = ['DATETIME']
    new_columns = cols_to_order + (df.columns.drop(cols_to_order).tolist())
    df = df[new_columns]
    agg.append(df)
```

```python
df = pd.concat(agg)
df.head(10)
dfs = [x for _, x in df.groupby('DATETIME')]
for df in tqdm(dfs):
    dt = str(df.iloc[0]['DATETIME'].split[' '])
    # print(dt)
    df.to_csv('CRN/'+dt+'.csv', index=False)
```
