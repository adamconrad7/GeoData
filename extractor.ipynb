```python
from bs4 import BeautifulSoup
import requests
# https://cdn.star.nesdis.noaa.gov/GOES18/ABI/CONUS/01/ - https://cdn.star.nesdis.noaa.gov/GOES18/ABI/CONUS/16/
# url = 'https://cdn.star.nesdis.noaa.gov/GOES18/ABI/CONUS/'
for i in range(1,17):
    url = 'https://cdn.star.nesdis.noaa.gov/GOES16/ABI/CONUS/'
    
    band = str(i)
    if i<10:
        band = '0'+band
    url += band + '/'
    print(url)
```

```python


response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
links = []
for a in soup.find_all('a'):

    link = a.get('href')
    # if '5000x3000' in link:
    if '416x250' in link:
    
        # print(link, a.next_element.next_element)
        links.append(url + link)
        # print(url + link)
    # print(a)
    # for ae in a.next_elements:
    #     print('\t', ae)

    # for link in soup.find_all('a'):
    # print(link.get('href'))
print(links[0])
total = len(links)
print(total)
```

```python
inc = 0
for link in links:
    
    if inc%10 == 0:
        print(inc/total*100)
    
        file_name = 'data/' + link.split(url)[1]
        # print(file_name)
        with open(file_name, "wb") as file:
           ###check if we have file already
            # get request
            response = requests.get(link)
            # write to file
            file.write(response.content)
    inc+=1
    
```
