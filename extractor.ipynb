```python
# https://cdn.star.nesdis.noaa.gov/GOES18/ABI/CONUS/01/ - https://cdn.star.nesdis.noaa.gov/GOES18/ABI/CONUS/16/
# url = 'https://cdn.star.nesdis.noaa.gov/GOES18/ABI/CONUS/'
for i in range(1,17):
    url = 'https://cdn.star.nesdis.noaa.gov/GOES18/ABI/CONUS/'
    
    band = str(i)
    if i<10:
        band = '0'+band
    url += band + '/'
    print(url)
```

```python
from bs4 import BeautifulSoup
import requests

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```
